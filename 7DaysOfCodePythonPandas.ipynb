# Day 2 - Limpeza de Dados

Os itens do acervo em uma biblioteca são organizados por um sistema de classificação de acordo com o respectivo tema. Existem diversos sistemas, mas este conjunto está de acordo com a **CDU - Classificação Decimal Universa**l. Esta classificação é decimal, pois varia de acordo com a classe de cada assunto:
* 000 a 099: Generalidades. Ciência e conhecimento.
* 100 a 199: Filosofia e psicologia.
* 200 a 299: Religião.
* 300 a 399: Ciências sociais.
* 400 a 499: Classe vaga. Provisoriamente não ocupada.
* 500 a 599: Matemática e ciências naturais.
* 600 a 699: Ciências aplicadas.
* 700 a 799: Belas artes.
* 800 a 899: Linguagem. Língua. Linguística.
* 900 a 999: Geografia. Biografia. História.
Vamos criar uma nova coluna com os dados da classe pertencente, de acordo com a coluna `localizacao`

## Adicionando coluna Classe - CDU

Verificando o tipo das colunas, para garantir que a coluna `localizacao` contem dados numéricos

```python
dados_sem_duplicidade.info()
```

Criando uma lista com as classes

```python
classes_cdu = []
```

Mapeando os valores de acordo com as condições

```python
for i in dados_sem_duplicidade['localizacao']:
  if 0 <= i <= 99:
    classes_cdu.append('Generalidades. Ciência e conhecimento.')
  elif 100<= i <= 199:
    classes_cdu.append('Filosofia e psicologia.')
  elif 200<= i <= 299:
    classes_cdu.append('Religião.')
  elif 300<= i <= 399:
    classes_cdu.append('Ciências sociais.')
  elif 400<= i <= 499:
    classes_cdu.append('Classe vaga. Provisoriamente não ocupada.')
  elif 500<= i <= 599:
    classes_cdu.append('Matemática e ciências naturais.')
  elif 600<= i <= 699:
    classes_cdu.append('Ciências aplicadas.')
  elif 700<= i <= 799:
    classes_cdu.append('Belas artes.')
  elif 800<= i <= 899:
    classes_cdu.append('Linguagem. Língua. Linguística.')
  elif 900<= i <= 999:
    classes_cdu.append('Geografia. Biografia. História.')
  else:
    classes_cdu.append('Sem classificação')
```

Adicionando a coluna `classe_CDU` no dataframe

```python
dados_sem_duplicidade['classe_CDU'] = classes_cdu
```

## Distribuição dos empréstismos pelos tipos de vínculo dos usuários

Desta forma, a diretoria poderá entender qual é o público que está utilizando a biblioteca e assim tomar decisões em continuar com a estratégia de negócio atual ou modificá-la.

Função para gerar tabela com a distribuição de dados categóricos com relação a um valor quantitativo, por exemplo: tipo de vínculo dos usúarios das bibliotecas com relação a quantidade de exemplares emprestados

```python
def distribuicao(df,label,values,label_name: str='',values_name: str='') -> pd.DataFrame:
  '''Função para gerar tabela com a distribuição de dados categóricos com relação a um valor quantitativo

  df: pd.DataFrame
    Dataframe em análise
  label: str
    String com o nome da coluna que contém as variaveis analisadas na distribuição
  values: str
    String com o nome da coluna que contém os valores da distribuição
  label_name: str
    String com o novo nome da coluna que contém as variaveis analisadas na distribuição, se o valor for vazio o nome não é alterado
  values_name: str
    String com o novo nome da coluna que contém as variaveis analisadas na distribuição, se o valor for vazio o nome não é alterado
  return = tabela : pd.DataFrame
    Tabela com a distribuição dos dados
  '''
  #Agrupando os dados da distribuição
  df_distribuicao = df.groupby(label,as_index = False)[[values]].count().sort_values(values,ascending = False).reset_index(drop = True)
  #Inserindo uma coluna com a porcentagem da variável categórica com relação ao todo
  df_distribuicao['%'] = round((df_distribuicao[values]/df_distribuicao[values].sum())*100,2)
  #Renomeando as colunas
  #Coluna das variáveis categóricas
  if label_name != '':
    df_distribuicao.rename(columns={label : label_name},inplace = True)
  #Coluna dos valores
  if values_name != '':
    df_distribuicao.rename(columns={values : values_name},inplace = True)
  return df_distribuicao
```

```python
emprestimos_vinculo = distribuicao(df = dados_sem_duplicidade,label = 'tipo_vinculo_usuario',values = 'id_exemplar',label_name = 'Vínculo do usuário', values_name = 'Quantidade de empréstimos')
emprestimos_vinculo
```

## Gráfico Box Plot por ano

```python
def boxplot(df,x: str='',y: str='',titulo: str='Box Plot',xlabel: str='',ylabel: str=''):
  '''
  Função para gerar gráfico Box Plot
  df: pd.DataFrame
    Dataframe que contém os dados
  x: str
    Nome da coluna do dataframe com os dados que ficarão no eixo x do gráfico
  y: str
    Nome da coluna do dataframe com os dados que ficarão no eixo y do gráfico
  titulo: str
    Título do gráfico
  xlabel: str
    Indicação a ser apresentada no eixo x do gráfico, se vazio usa o nome da coluna
  ylabel: str
    Indicação a ser apresentada no eixo x do gráfico, se vazio usa o nome da coluna
  '''
  #Verificando label dos eixos
  if xlabel == '':
    xlabel == x
  if ylabel == '':
    ylabel == y
  fig = px.box(df,
             x = x,
             y = y,
             labels = {x : xlabel,y : ylabel},
             title = titulo)
  fig.update_xaxes(tickmode='linear')
  fig.show()
```

## Leitura do arquivo de previsão de empréstimos 2022

```python
url_previsao_2022 = 'https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_7-Apresentando_resultados_em_HTML/Dataset/previsao'
previsao_2022 = pd.read_csv(url_previsao_2022,sep = " ",names = ['curso','2022'],skiprows = 1)
previsao_2022
```

## Mapeando dicionário com os nomes dos cursos

```python
nomes_cursos = {'administracao' : 'ADMINISTRAÇÃO',
                'arquitetura_e_urbanismo' : 'ARQUITETURA E URBANISMO',
                'artes_cenicas' : 'ARTES CÊNICAS',
                'bioinformatica' : 'BIOINFORMÁTICA',
                'bioquimica' : 'BIOQUÍMICA',
                'ciencia_e_engenharia_de_materiais' : 'CIÊNCIA E ENGENHARIA DE MATERIAIS',
                'ciencia_tecnologia_e_inovacao' : 'CIÊNCIA, TECNOLOGIA E INOVAÇÃO',
                'ciencias_da_saude': 'CIÊNCIAS DA SAÚDE',
                'ciencias_odontologicas' : 'CIÊNCIAS ODONTOLÓGICAS',
                'design' : 'DESIGN',
                'engenharia_civil' : 'ENGENHARIA CIVIL',
                'engenharia_software' : 'ENGENHARIA DE SOFTWARE',
                'filosofia' : 'FILOSOFIA',
                'gestao_da_informacao_e_do_conhecimento' : 'GESTÃO DA INFORMAÇÃO E DO CONHECIMENTO'
                }
```

```python
previsao_2022.curso = previsao_2022.curso.map(nomes_cursos)
previsao_2022
```

Renomeando coluna 2022 para tipo int, conforme os dados de empréstimos dos anos 2017-2019

```python
previsao_2022.rename(columns = {'2022' : 2022}, inplace = True)
```

## Unindo os data frame de empréstimos ano da pós-graduação, com a previsão de 2022

```python
emprestimos_por_ano_pg_df = pd.merge(left= emprestimos_por_ano_pg, right = previsao_2022[[2022,'curso']], on = 'curso')
emprestimos_por_ano_pg_df
```

## Diferenças percentuais

```python
emprestimos_por_ano_pg_df['2017-2018'] = round(((emprestimos_por_ano_pg_df[2018] - emprestimos_por_ano_pg_df[2017])/emprestimos_por_ano_pg_df[2017])*100,2)
emprestimos_por_ano_pg_df['2018-2019'] = round(((emprestimos_por_ano_pg_df[2019] - emprestimos_por_ano_pg_df[2018])/emprestimos_por_ano_pg_df[2018])*100,2)
emprestimos_por_ano_pg_df['2019-2022'] = round(((emprestimos_por_ano_pg_df[2022] - emprestimos_por_ano_pg_df[2019])/emprestimos_por_ano_pg_df[2019])*100,2)
emprestimos_por_ano_pg_df
```

## Ajustes descrição da tabela

Nomes dos cursos com apenas a primeira leita maiúscula

```python
emprestimos_por_ano_pg_df['curso'] = emprestimos_por_ano_pg_df['curso'].str.capitalize()
```

Renomeando coluna 'curso'

```python
emprestimos_por_ano_pg_df.rename(columns = {'curso' : 'Curso'},inplace = True)
```

Removendo as colunas com o total dos anos

```python
emprestimos_por_ano_pg_df.drop(columns = [2017,2018,2019,2022], axis = 'columns',inplace = True)
```

Inserindo % nos valores de variação

```python
tab_dif_emprest_curso = emprestimos_por_ano_pg_df.style.format('{:} %',subset = ['2017-2018','2018-2019','2019-2022'])
```

Alterando a cor dos números

```python
tab_dif_emprest_curso.text_gradient(cmap = 'RdYlGn',vmin = 0, vmax = 0.1, axis = 'columns',subset = ['2017-2018','2018-2019','2019-2022'])
```
